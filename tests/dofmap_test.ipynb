```python
# add ../product_fem to path
import sys
sys.path.append(sys.path[0] + '/..')

import numpy as np
from fenics import *
from product_fem import product_fem as pf
```

```python
# test using product function f(x,y) = xy
f = lambda x,y: x*y
n = 11
mesh1 = UnitIntervalMesh(n-1)
V1 = FunctionSpace(mesh1, 'CG', 1)
W = pf.ProductFunctionSpace(V1)
d2c = W.dofmap._dofs_to_coords # dict {dof: (x, y)}
assert len(d2c) == n**2
```

```python
f_vec = []
for xy in d2c.values():
    f_vec.append(f(xy[0], xy[1]))
print(f_vec)
```

```python
mesh2 = UnitSquareMesh(n-1, n-1)
V2 = FunctionSpace(mesh2, 'CG', 1)
dof_coords = V2.tabulate_dof_coordinates()
dof_coords_dict = dict()
for i, xy in enumerate(dof_coords):
    dof_coords_dict[i] = tuple(xy)
```
